```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pend(y, t):
    x, xdot = y
    ydot = [xdot, - np.sin(x)]
    return ydot
```

```python
batches=128
trajectories=15000
timesteps=51
```

```python
initial_el=2 *(np.pi-0.01) *(np.random.rand(batches, trajectories)-0.5) #128, 15000
initial_vel=2 *(np.pi-0.01) *(np.random.rand(batches, trajectories)-0.5)
#initial_el = (np.pi-0.01) *(np.random.rand(batches, trajectories))
# v_max=sqrt(2(H-cosx_01))
#initial_vel=np.multiply(np.sqrt(2*(1+np.cos(initial_el))),(2*np.random.rand(batches, trajectories)-1))
#initial_vel = np.cos(initial_el)
#initial_vel=0*initial_el
initial=np.stack((initial_el,initial_vel), axis=-1)
#np.shape(initial)
plt.scatter(initial[0,:,0],initial[0,:,1])
```

```python
t = np.linspace(0, 10, timesteps)
```

```python
from scipy.integrate import odeint
```

```python
data=np.zeros((batches, trajectories, timesteps, 2))
for i in range(batches):
    print(i)
    for j in range(trajectories):
        data[i,j,:,:]=odeint(pend,initial[i,j,:], t)
```
